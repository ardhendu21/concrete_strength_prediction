```python
#cleaning outlier
def clean_outliers(data):
    for col in data.columns[0:-1]:
        Q1= data[col].quantile(0.25)
        Q3= data[col].quantile(0.75)
        IQR=Q3-Q1
        c1=Q1-(1.5*IQR)
        c2=Q3+(1.5*IQR)
        data.loc[data[col] < c1, col] = c1
        data.loc[data[col] > c2, col] = c2
    return data
```

```python
# Removing outliers
data=clean_outliers(data)
```

```python
# check for outliers
features=data.columns
plt.figure(figsize=(18,7))
sns.boxplot(data=data[features[0:-1]]);
```

```python
# splitting and scaling the dataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X = data.iloc[:,:-1] 
y = data.iloc[:,-1] 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2) 
sc = StandardScaler() 
X_train = sc.fit_transform(X_train) 
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression  
from sklearn.metrics import mean_squared_error   
from sklearn import metrics
lm = LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
m1=model.score(X_test, y_test)
RMSE1=np.sqrt(metrics.mean_squared_error(y_test, predictions))
print('Liner Regression Model Score', model.score(X_test, y_test))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, interaction_only=True)
X_train_ = poly.fit_transform(X_train)
X_test_ = poly.fit_transform(X_test)
poly_clf = LinearRegression()
poly_clf.fit(X_train_, y_train)
predictions2 = poly_clf.predict(X_test_)
pf_score = poly_clf.score(X_test_, y_test)
print("2D Polynomial Model Score:",pf_score)
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predictions2)))
```

```python
#Model 3 Polynomial Regressor with Degree=3
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3, interaction_only=True)
X_train__ = poly.fit_transform(X_train)
X_test__= poly.fit_transform(X_test)
poly_clf = LinearRegression()
poly_clf.fit(X_train__, y_train)
predictions3 = poly_clf.predict(X_test__)
pf_score = poly_clf.score(X_test__, y_test)
print("3D Polynomial Model Score:",pf_score)
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predictions3)))
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train , y_train)
y_pred = dt_model.predict(X_test)
acc_DT=metrics.r2_score(y_test, y_pred)
print('Decision Tree Regressor Score: ',acc_DT)
print('Root Mean Squared Error: ',np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
